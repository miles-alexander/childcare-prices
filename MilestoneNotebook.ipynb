```python
import pandas as pd
```

```python
import numpy as np
```

```python
childcare_prices = pd.read_excel('nationaldatabaseofchildcareprices.xlsx')
```

```python
childcare_prices
```

```python
# find if there are missing values
any_missing = childcare_prices.isnull().values.any()
print(any_missing)
```

```python
# display rows with missing values
childcare_prices[childcare_prices.isnull().any(axis = 1)]
```

```python
# list of columns to impute missing values (columns we are focusing on)
columns_to_impute = ['MCInfant', 'MCToddler', 'MCPreschool', 'MFCCInfant', 'MFCCToddler', 'MFCCPreschool', 'MHI', 'ME']

# loop through each column in the list to apply imputation
for column in columns_to_impute:
    childcare_prices[column] = childcare_prices[column].fillna(childcare_prices[column].mean())  # replace nulls with the overall mean
    childcare_prices[column] = childcare_prices[column].fillna(childcare_prices.groupby('State_Name')[column].transform('mean'))  # replace remaining nulls with state-wise mean
```

```python
childcare_prices.to_excel('cleaned_childcare_data.xlsx', index = False)
```
